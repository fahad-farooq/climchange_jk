```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('./datasets/ceda/tmn/tmn5.csv',index_col='DATE',parse_dates=True)
data.reset_index(inplace=True)
data['MONTH'] = data['DATE'].dt.month
data['DATE'] = data['DATE'].dt.year
data.rename(columns={'DATE':'YEAR'},inplace=True)
data=data[['YEAR','MONTH','AVG MIN TEMPERATURE']]
data=pd.DataFrame(data)
data.set_index('YEAR',inplace=True)
data.head()
```

```python
sns.set_palette(['green'])
data_jan=data.loc[data["MONTH"]==1]
plt.figure(figsize=(16,9))
plt.title("Changes in average minimum temperature for January over the years: 1901-2019")
sns.lineplot(data=data_jan['AVG MIN TEMPERATURE'],label='Year')
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.savefig("tmn_jan.png")
```

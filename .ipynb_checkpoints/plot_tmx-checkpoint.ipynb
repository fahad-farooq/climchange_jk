```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('./datasets/ceda/tmx/tmx5.csv',index_col='DATE',parse_dates=True)
data.reset_index(inplace=True)
data['MONTH'] = data['DATE'].dt.month
data['DATE'] = data['DATE'].dt.year
data.rename(columns={'DATE':'YEAR'},inplace=True)
data=data[['YEAR','MONTH','AVG MAX TEMPERATURE']]
data=pd.DataFrame(data)
data.set_index('YEAR',inplace=True)
data.head()
```

```python
sns.set_palette(['red'])
data_jan=data.loc[data["MONTH"]==1]
plt.figure(figsize=(16,9))
plt.title("Changes in average maximum temperature for January over the years: 1901-2019")
sns.lineplot(data=data_jan['AVG MAX TEMPERATURE'],label='Year')
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.savefig("tmx_jan.png")
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('./datasets/ceda/cld/cld5.csv',index_col='DATE',parse_dates=True)
data.reset_index(inplace=True)
data['MONTH'] = data['DATE'].dt.month
data['DATE'] = data['DATE'].dt.year
data.rename(columns={'DATE':'YEAR'},inplace=True)
data=data[['YEAR','MONTH','CLOUD COVER']]
data=pd.DataFrame(data)
data.set_index('YEAR',inplace=True)
data.head()
```

```python
data_jan=data.loc[data["MONTH"]==1]
plt.figure(figsize=(16,9))
plt.title("Changes in cloud cover for January over the years: 1901-2019")
sns.lineplot(data=data_jan['CLOUD COVER'],label='Year')
plt.xlabel("Year")
plt.ylabel("Cloud Cover")
plt.savefig("cld_jan.png")
```
